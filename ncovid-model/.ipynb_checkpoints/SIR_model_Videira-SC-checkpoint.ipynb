```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Total population of Videira-SC, N.
N = 47510
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1./10 
# A grid of time points (in days)
t = np.linspace(0, 160, 160)
```

```python
# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt
```

```python
# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T
```

```python
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w', figsize=(12,8))
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, S, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Population')
#ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=4)
ax.xaxis.set_tick_params(length=4)
#ax.grid(b=True, which='major', c='w', lw=2, ls='-')
ax.grid(True)
legend = ax.legend()
legend.get_frame().set_alpha(1)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

```python
# Plot the the curve for infected population
fig = plt.figure(facecolor='w', figsize=(12,8))
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, I, 'r', alpha=0.5, lw=2, label='Infected')
ax.set_xlabel('Time /days')
ax.set_ylabel('Population infected')
#ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=4)
ax.xaxis.set_tick_params(length=4)
#ax.grid(b=True, which='major', c='w', lw=2, ls='-')
ax.grid(True)
legend = ax.legend()
legend.get_frame().set_alpha(1)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

```python
pop_acometida = (I.max()*100)/N
pop_acometida
```

```python
print ('A estimativa para população acometida, no pico maximo, usando valor de beta', beta,
       'a partir do primeiro caso confirmado é de',
       pop_acometida, '%, contabilizando', I.max(), 'habitantes.')
```

```python
# Plot cumulative cases
cumulative_cases = np.cumsum(I, dtype=float)
fig = plt.figure(facecolor='w', figsize=(12,8))
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, cumulative_cases, 'r', alpha=0.5, lw=2, label='Infected')
ax.set_xlabel('Time /days')
ax.set_ylabel('Population infected')
#ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=4)
ax.xaxis.set_tick_params(length=4)
#ax.grid(b=True, which='major', c='w', lw=2, ls='-')
ax.grid(True)
legend = ax.legend()
legend.get_frame().set_alpha(1)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
