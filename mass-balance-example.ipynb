```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Input parameters
# Initial biomass concentration
X0= 2.0 #g/L
# Initial substrate concentration
S0 = 10.0 #g/L
# Rate mu_max
mu_max = 0.02
#Half saturation constant (mg/L)
Ks = 0.03
#Endogenous decay rate constant, h-1
kd = 0.002
#Yield coeff (mass biomass(mg/L)/mass substrate utilized(mg/L))
Y = 0.8
# Timesteps (days)
t = np.linspace(0, 90, 90)
```

```python
# The model differential equations.
def deriv(y, t, mu_max, Ks, kd, Y):
    X, S = y
    dXdt = ((mu_max*S*X)/(Ks+S))- (kd*X)
    dSdt = -(1/Y)*((mu_max*S*X)/(Ks+S))
    return dXdt, dSdt
# Initial conditions vector
y0 = X0, S0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(mu_max, Ks, kd, Y))
X,S = ret.T
```

```python
# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w', figsize=(12,8))
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, X, 'b', alpha=0.5, lw=2, label='Biomass')
ax.plot(t, S, 'r', alpha=0.5, lw=2, label='Substrate')
ax.set_xlabel('Time /hrs')
ax.set_ylabel('Concentration')
ax.yaxis.set_tick_params(length=4)
ax.xaxis.set_tick_params(length=4)
ax.grid(True)
legend = ax.legend()
legend.get_frame().set_alpha(1)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```
